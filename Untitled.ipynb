```python


import datetime

dt = datetime.datetime(2000, 1, 1)
end = datetime.datetime(2004, 12, 31)
step = datetime.timedelta(days=1)

result = []

year = []
month = []
day = []

while dt < end:
    result.append(dt.strftime('%Y-%m-%d %H:%M:%S'))
    year.append(dt.year)
    month.append(dt.month)
    day.append(dt.day)
    dt += step

print(year)
print(month)
print(day)
```
